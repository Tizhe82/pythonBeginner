```python
#pyramid
def pattern(n):
    k = 2 * n - 2
    for i in range(0,n):
        for j in range(0,k):
            print(end=" ")
        k = k - 1
        for j in range(0, i + 1):
            print("* ", end="")
            print("\n")
pattern(5)
```

```python
#inverse pyramid
def pattern(n):
    k = 2*n - 2
    for i in range(n, -1,-1):
        for j in range(k,0,-1):
            print(end=" ")
            k = k + 1
            for j in range(0, i + 1):
                print("* ", end="")
            print("\r")
pattern(10)
```

```python
#right start pattern
def pattern(n):
    for i in range(0,n):
        for j in range(0,i + 1):
            print("* ", end="")
        print("\r")
        for i in range(n,0,-1):
            for j in range(0,i + 1):
                print("* ", end="")
                print("\r")
pattern(10)
```
